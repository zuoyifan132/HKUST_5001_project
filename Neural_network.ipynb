# Data Preprocessing

```python
y = data['Revenue']
X = data.drop(["Revenue"], axis=1)
```

### Data splitting

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.3, random_state=0, stratify=y)
```

```python
print("X_train : " + str(X_train.shape))
print("X_test : " + str(X_test.shape))
print("y_train : " + str(Y_train.shape))
print("y_test : " + str(Y_test.shape))
X_train.info()
```

### Check null value

```python
np.sum(pd.isnull(X_train))
```

### Normalization with Training data set

```python
skewness = X_train.skew(axis=0)
skewness[(skewness.iloc[:] > 0.5) | (skewness.iloc[:] < -0.5)]
```

```python
need_transform = ["Administrative", "Administrative_Duration", "Informational", "Informational_Duration", "ProductRelated",
                 "ProductRelated_Duration", "PageValues", "OperatingSystems", "Browser",
                 "Region", "TrafficType"]

X_train[need_transform] = np.log1p(X_train[need_transform])
skewness = X_train.skew(axis=0)
skewness[(skewness.iloc[:] > 0.5) | (skewness.iloc[:] < -0.5)]
```

```python
X_train.info()
```

### Normalization with Training data set

```python
need_transform = ["Administrative", "Administrative_Duration", "Informational", "Informational_Duration", "ProductRelated",
                 "ProductRelated_Duration", "PageValues", "OperatingSystems", "Browser",
                 "Region", "TrafficType"]

X_test[need_transform] = np.log1p(X_test[need_transform])
skewness = X_test.skew(axis=0)
skewness[(skewness.iloc[:] > 0.5) | (skewness.iloc[:] < -0.5)]
```

### Converting Categorial features

```python
X_train = pd.get_dummies(X_train)
X_test = pd.get_dummies(X_test)
```

```python
X_train
```

### Converting boolean to 0 and 1

```python
X_train["Weekend"] =X_train["Weekend"].map({True:1 , False :0})
X_test["Weekend"] =X_test["Weekend"].map({True:1 , False :0})
Y_train = Y_train.map({True:1 , False :0})
Y_test = Y_test.map({True:1 , False :0})
```

```python
print("X_train : " + str(X_train.shape))
print("X_test : " + str(X_test.shape))
print("y_train : " + str(Y_train.shape))
print("y_test : " + str(Y_test.shape))
```

### Standardization and Rescaling

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_train)
trans_X_train = scaler.transform(X_train)
X_train = pd.DataFrame(trans_X_train, columns=X_train.columns)

scaler.fit(X_test)
trans_X_test = scaler.transform(X_test)
X_test = pd.DataFrame(trans_X_test, columns=X_test.columns)
```

# Neural Network Training

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score

model = MLPClassifier(solver='adam', hidden_layer_sizes=(6, 3), activation='relu', alpha=0.0001, max_iter=1000)

model.fit(X_train, Y_train)
y_pred = model.predict(X_test)
weights = model.coefs_
```

```python
# accuracy
accuracy_score(Y_test, y_pred)
```

```python
from sklearn.metrics import f1_score

# f1 score
print(f1_score(Y_test, y_pred, average='macro'))
```

```python
feature_weights = np.array(weights[0])
feature_weights_mean = feature_weights.mean(axis=1)
feature_rank = pd.DataFrame(feature_weights_mean, index=X_train.columns.tolist())
feature_rank
```
